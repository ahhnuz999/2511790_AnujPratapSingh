```python
#Which passenger had the highest fare paid relative to their age?
df['fare_per_year'] = df['Fare'] / df['Age']
high_fare_age = df[df['fare_per_year'] > 5]
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)
result = high_fare_age_srt[['Name', 'fare_per_year']]
result.head(30)
```

```python
# Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their class?
df['fare_per_class'] = df['Fare'] / df['Pclass']
adult_males = df[(df['Sex']=='male') & (df['Age']>=18)]
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)
result = adult_males_srt[['Name','Age','fare_per_class']]
result.head()
```

```python
total_fare = df['Fare'].sum()
print("Total Fare =",total_fare)
fare_one = df[df['Pclass']==1]['Fare'].sum()
fare_two = df[df['Pclass']==2]['Fare'].sum()
fare_three = df[df['Pclass']==3]['Fare'].sum()
print("Total Fare 1st class =",fare_one)
print("Total Fare 2nd class =",fare_two)
print("Total Fare 3rd class =",fare_three)
```

```python
fare_list = df.groupby('Pclass')['Fare'].sum().tolist()
print("Fare in list",fare_list)
```

```python
revenue = (fare_list)/(total_fare)
print(revenue)
```

```python
#What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?
df['age_group'] = pd.cut(df['Age'], bins=[0, 18, 65, 200], labels = ['child', 'adult', 'senior'], right=False)

total_passengers = df.shape[0]

child_passengers = df[df['age_group'] == 'child'].shape[0]
adult_passengers = df[df['age_group'] == 'adult'].shape[0]
senior_passengers = df[df['age_group'] == 'senior'].shape[0]

child_passengers_percentage = (child_passengers/total_passengers) * 100
adult_passengers_percentage = (adult_passengers/total_passengers) * 100
senior_passengers_percentage = (senior_passengers/total_passengers) * 100

print(child_passengers_percentage)
print(adult_passengers_percentage)
print(senior_passengers_percentage)
```
